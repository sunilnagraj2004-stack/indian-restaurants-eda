```python
df = pd.read_csv(r"C:\Users\Sunil\Downloads\data\indian_restaurants_details_cleaned_data.csv")
df.head()
```

```python
print(df.shape)
print(df.columns.tolist())
```

```python
print(df.isnull().sum())
df = df.dropna(subset=['rating', 'city', 'cusine'])
print("Cleaned shape:", df.shape)
```

```python
plt.figure(figsize=(12,5))
df['city'].value_counts().head(10).plot(kind='bar', color='steelblue')
plt.title('Top 10 Cities with Most Restaurants in India')
plt.xlabel('City')
plt.ylabel('Number of Restaurants')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12,5))
df['cusine'].value_counts().head(10).plot(kind='bar', color='coral')
plt.title('Top 10 Most Popular Cuisines in India')
plt.xlabel('Cuisine')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6,5))
df['online_order'].value_counts().plot(kind='bar', color=['green','red'])
plt.title('Online Order Availability')
plt.xlabel('Online Order')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12,5))
df.groupby('city')['rating'].mean().sort_values(ascending=False).head(10).plot(kind='bar', color='purple')
plt.title('Top 10 Cities by Average Restaurant Rating')
plt.xlabel('City')
plt.ylabel('Average Rating')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6,5))
df['table_reservation'].value_counts().plot(kind='pie', autopct='%1.1f%%', colors=['lightblue','orange'])
plt.title('Table Reservation Availability')
plt.ylabel('')
plt.show()
```
